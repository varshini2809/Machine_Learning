```python
x=df.drop("Y house price of unit area",axis=1)
x
```

```python
y=df['Y house price of unit area']
y
```

```python
polynomial_reg=PolynomialFeatures(degree=2,include_bias=False)
```

```python
poly_feat=polynomial_reg.fit_transform(x)
poly_feat.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(poly_feat,y,test_size=0.3,random_state=102)
```

```python
polynomial=LinearRegression()
```

```python
polynomial.fit(x_train,y_train)
```

```python
y_pred=polynomial.predict(x_test)
```

```python
a=pd.DataFrame({'y_test':y_test,'y_pred':y_pred,"residuals":y_test-y_pred}) #incomplete
a
```

```python
poly_ABS=metrics.mean_absolute_error(y_test,y_pred)
poly_MSE=metrics.mean_squared_error(y_test,y_pred)
poly_RSE=np.sqrt(poly_MSE)
```
