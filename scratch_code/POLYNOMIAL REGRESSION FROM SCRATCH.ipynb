```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.array([1,2,3,4,5])
y=np.array([2.1,3.8,6.5,9.2,11.3])
```

```python
degree=2
```

```python
x_poly=np.column_stack([x**i for i in range(degree+1)]) # Creating design matrix
x_poly
```

```python
weights=np.zeros(degree+1)
weights
```

```python
lr=0.001
epochs=10000
```

```python
for _ in range(epochs):
  #calculate prediction
  predictions=np.dot(x_poly,weights) # dot->multiplication
  #calculate error
  error=predictions-y
  #update the weight
  gradient=np.dot(x_poly.T,error)/len(x)  # .T -> transpose
  weights-=lr*gradient
 # print(f'weights:{_}:{weights}')
```

```python
def predict(x,weights):
  x_poly=np.array([x**i for i in range(degree+1)])
  return np.dot(x_poly,weights)
```

```python
plt.scatter(x,y,label="Data Points")
x_range=np.linspace(min(x),max(x),100)  #linspace -> all values are eventually spaced
y_pred=[predict(x,weights) for x in x_range]
plt.plot(x_range,y_pred,color='r',label="polynomial Regression")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
