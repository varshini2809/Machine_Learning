```python
df['Gender']=df['Gender'].map({"Male":1,"Female":0})
df['Married']=df['Married'].map({"Yes":1,"No":0})
df['Education']=df['Education'].map({"Graduate":1,"Not Graduate":0})
df['Dependents'].replace('3+',3,inplace=True)
df['Self_Employed']=df['Self_Employed'].map({"Yes":1,"No":0})
df['Property_Area']=df['Property_Area'].map({"Semiurban":1,"Urban":2,"Rural":3})
df['Loan_Status']=df['Loan_Status'].map({"Y":1,"N":0})
```

```python
df['Loan_Status']
```

```python
df['Gender']=df['Gender'].mean()
df['Married']=df['Married'].mean()
df['Dependents']=df['Dependents'].mode()
df['Self_Employed']=df['Self_Employed'].mean()
df['LoanAmount']=df['LoanAmount'].mean()
df['Loan_Amount_Term']=df['Loan_Amount_Term'].mean()
df['Credit_History']=df['Credit_History'].mean()

df['Property_Area']=df['Property_Area'].mean()
df['Loan_Status']=df['Loan_Status'].mean()
```

```python
df.isnull().sum()
```

```python
df['Loan_Status']
```

```python
x=df.drop(columns=['Loan_ID','Loan_Status','Dependents']).values
y=df['Loan_Status'].values
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.preprocessing import StandardScaler
sta=StandardScaler()
x_train=sta.fit_transform(x_train)
x_test=sta.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
rfc=RandomForestClassifier(criterion='entropy',random_state=42)
rfc.fit(x_train,y_train)
```

```python
y_pred_train=rfc.predict(x_train)
print(classification_report(y_train,y_pred_train))
```

```python
y_pred_test=rfc.predict(x_test)
print(classification_report(y_test,y_pred_test))
```
