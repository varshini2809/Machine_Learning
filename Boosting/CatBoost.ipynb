```python
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
```

```python
x = train.iloc[:, :-1]
y = train.iloc[:, -1]
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(x,y, test_size=0.25, random_state=28)
```

```python
# Define the hyperparameters for the CatBoost algorithm
params = {'learning_rate': 0.1, 'depth': 6,'l2_leaf_reg': 3, 'iterations': 5}

# Initialize the CatBoostClassifier object
# with the defined hyperparameters and fit it on the training set
model = CatBoostClassifier(**params)
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_val)
accuracy = accuracy_score(Y_val, y_pred) * 100
print("Accuracy  ",accuracy)
```
