```python
import numpy as np
import pandas as pd
```

```python
import xgboost as xgb
```

```python
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/datasets_228_482_diabetes.csv")
```

```python
x=df.iloc[:,:-1].values
```

```python
y=df.iloc[:,-1:].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8)
```

```python
model=XGBClassifier(learning_rate=0.076,gamma=0.2,random_state=4)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
accuracy=accuracy_score(y_test,y_pred)*100
accuracy
```

```python
model_xgb=XGBClassifier()
from sklearn.model_selection import GridSearchCV
param_grid={
    'learning_rate':[0.01,0.25,0.5],
    'max_depth':[3,4,5],
    'subsample':[0.9,0.8,1.0],
    'colsample_bytree':[0.5,0.6,0.7],
    'n_estimators':[500,600,700]
}
grid_search=GridSearchCV(estimator=model_xgb,param_grid=param_grid,scoring='accuracy',cv=5,verbose=1)
```

```python
grid_search.fit(x_train,y_train)
print("Best parameter",grid_search.best_params_)
print("Best acuracy",grid_search.best_score_)
```
